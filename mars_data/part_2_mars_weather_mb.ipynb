```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Extract all rows of data
all_rows = soup.find_all('tr')
print(all_rows)
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list for table data and table headers
table_data = []
table_headers = []

# Loop through the scraped data to create a list of rows
for row in all_rows: 
    
    # Extract headers from dataset
    headers = row.find_all('th')

    # Loop through headers to collect into a list
    for header in headers:
        name = [header.get_text(strip = True) for header in headers]
        # Append to list
        table_headers.append(name)
        # Break after first iteration to collect only once
        break
    # Extract columns from dataset
    columns = row.find_all('td')

    # loop through columns to collect each row's data into lists
    for column in columns: 
        data = [column.get_text(strip = True) for column in columns]
        # Append to list
        table_data.append(data)
        # Break after first iteration to collect only once
        break
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
mars_df = pd.DataFrame(table_data, columns = table_headers)

mars_df
```

```python
# Confirm DataFrame was created successfully
mars_df.head()
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
mars_df.dtypes
```

```python
# Change data types for data analysis

# List out columns to separate headers from data
mars_df.columns = ['id', 'terrestrial_date', 'sol', 'ls', 'month', 'min_temp', 'pressure']

# Change to datetime
mars_df.terrestrial_date = pd.to_datetime(mars_df.terrestrial_date)

# Change to Int
mars_df['sol'] = mars_df['sol'].astype('int32')
mars_df['ls'] = mars_df['ls'].astype('int32')
mars_df['month'] = mars_df['month'].astype('int32')

# Change to Float
mars_df['min_temp'] = mars_df['min_temp'].astype('float64')
mars_df['pressure'] = mars_df['pressure'].astype('float64')
```

```python
# Confirm type changes were successful by examining data types again
mars_df.dtypes
```

Analyze your dataset by using Pandas functions to answer the following questions:

```python
# 1. How many months are there on Mars?
months = mars_df['month'].max()
months
```

```python
# 2. How many sols (Martian days) worth of data are there?
sols = mars_df['sol'].count()
sols
```

```python
# 3. What is the average minimum temperature by month?
avg_min_temp = mars_df.groupby('month')['min_temp'].mean()
avg_min_temp
```

```python
# Plot the average minimum temperature by month
min_temp_chart = avg_min_temp.plot(kind = 'bar', title = 'Average Minimum Temperature by Month')

# Label axes
min_temp_chart.set_ylabel('Temperature in Celsius')
min_temp_chart.set_xlabel('Month')

# Format
plt.tight_layout
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location by sorting the previous graph
sorted_temp = avg_min_temp.sort_values(axis = 0, ascending = True)

sorted_temp_chart = sorted_temp.plot(kind = 'bar', title = 'Average Minimum Temperature by Month')

# Label axes
sorted_temp_chart.set_ylabel('Temperature in Celsius')
sorted_temp_chart.set_xlabel('Month')

# Format
plt.tight_layout
plt.show()
```

```python
# 4. What is the average pressure by month?
avg_pressure = mars_df.groupby('month')['pressure'].mean()
avg_pressure
```

```python
# Plot the average pressure by month
avg_pressure_chart = avg_pressure.plot(kind = 'bar', title = 'Average Pressure by Month')

# Label axes
avg_pressure_chart.set_ylabel('Atmospheric Pressure')
avg_pressure_chart.set_xlabel('Month')

# Format
plt.tight_layout
plt.show()
```

```python
# Identify the lowest and highest pressure months in Curiosity's location by sorting the previous graph
sorted_pressure = avg_pressure.sort_values(axis = 0, ascending = True)

sorted_pressure_chart = sorted_pressure.plot(kind = 'bar', title = 'Average Pressure by Month')

# Label axes
min_temp_chart.set_xlabel('Temperature in Celsius')
min_temp_chart.set_ylabel('Month')

# Format
plt.tight_layout
plt.show()
```

This cycle, especially if the pattern repeats year over year, would give you an idea of when each season is and during which months there may be a significant shift in weather. 



#### Atmospheric Pressure

The maximum atmospheric pressure on Mars is during month 9, and the minimum atmospheric pressure is during month 6. Looking at the Average Minimum Temperature Graph at the same time, the Atmospheric Pressure changes do not perfectly mirror the Temperature Changes. This leads one to believe that there may not be a correlation coefficent of 1 between pressure and temperature, and you may not be able to use one to predict the other. 

#### Year Length

Given the cycle that we can see on the Minimum Temperature over Time chart, we can see the same weather pattern occurring about a 600-700 Earth Days apart. This leads one to believe that the Martian Year is equivalent to between 600-700 Earth Days. This is also supported by the table scraped from the website. The first day recorded (month 6) is on 2012-08-16, and the next beginning of the 6th month is recorded on 2014-06-23, which is less than 2 years apart but more than a year and a half. 

This means that Mars orbits the sun slower than Earth, but it is still having cyclical seasons (minimum temperature) and atmospheric pressure.
